```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'datetime' is in the correct datetime format
e_data['datetime'] = pd.to_datetime(e_data['datetime'])
data['datetime'] = pd.to_datetime(data['datetime'])

# Filter the data to include only rows where the magnitude (MAG) is greater than 6
filtered_e_data = e_data[e_data['MAG'] > 3]

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Magnitude (MAG) on the first y-axis for magnitudes > 6
ax1.plot(filtered_e_data['datetime'], filtered_e_data['MAG'], label='Magnitude (MAG)', color='blue', linestyle='-', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Magnitude', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Format x-axis ticks for better readability
ax1.xaxis.set_major_locator(mdates.AutoDateLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Create a second y-axis for animal movement
ax2 = ax1.twinx()
ax2.plot(data['datetime'], data['sheep_sum_mean'], label='Sheep Movement', color='purple', linestyle='-', marker='x')
ax2.set_ylabel('Animal Movement', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add legends for both y-axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Add title and grid
plt.title('Date vs Magnitude (MAG > 6) and Sheep Movement')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'datetime' is in the correct datetime format
e_data['datetime'] = pd.to_datetime(e_data['datetime'])
data['datetime'] = pd.to_datetime(data['datetime'])

# Filter the data to include only rows where the magnitude (MAG) is greater than 4
filtered_e_data = e_data[e_data['MAG'] > 3]

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Magnitude (MAG) on the first y-axis for magnitudes > 4
ax1.plot(filtered_e_data['datetime'], filtered_e_data['MAG'], label='Magnitude (MAG)', color='blue', linestyle='-', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Magnitude', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Adjust the x-axis ticks to show all the dates
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=10))  # Adjust the interval as needed
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)  # Rotate the date labels for better visibility

# Create a second y-axis for animal movement
ax2 = ax1.twinx()
ax2.plot(data['datetime'], data['sheep_sum_mean'], label='Sheep Movement', color='purple', linestyle='-', marker='x')
ax2.set_ylabel('Animal Movement', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add legends for both y-axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Add title and grid
plt.title('Date vs Magnitude (MAG > 4) and Sheep Movement')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
